```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("employee.csv")
df
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="Gender",y="Age",data=df,color="pink",edgecolor="red")
plt.title("Gender vs Age")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df["Age"],bins=30,color="violet",kde=False)
plt.show()
```

```python
#scatter plot
sns.scatterplot(x="Age",y="ExperienceInCurrentDomain",data=df,hue=df["City"])
plt.show()
```
